```python
import urllib2
from bs4 import BeautifulSoup
import time
from time import sleep
import pandas as pd
```

```python
job_title = []
company_name = []
location = []
summary = []
for i in range(0,501,10):
    time.sleep(1)
    print 'starting loop', i
    web_page = 'http://www.indeed.com/jobs?q=data+scientist+$55,000&l=Chicago,+IL&start=' + str(i)
    print web_page
    soup55 = BeautifulSoup(urllib2.urlopen(web_page).read())
    for i in soup55('div', {'class': ' row result'}):
        for j in i('a', {'data-tn-element': 'jobTitle'}):
            print 'Job title is %s' % j.text
            job_title.append((j.text).encode('utf-8').strip())
        for j in i('span', {'class': 'company'}):
            print 'Comapany name is %s' % j.text 
            company_name.append((j.text).encode('utf-8').strip())
        for j in i('span', {'class': 'location'}):
            print 'Job is located in %s' % j.text
            location.append((j.text).encode('utf-8').strip())
        for j in i('span', {'class': 'summary'}):
            print 'Job summary is %s' % j.text
            summary.append((j.text).encode('utf-8').strip())
            
salary_55k = {'Job Title' : job_title, 'Company Name' : company_name, 
               'Location' : location, 'Summary' : summary, 'Salary' : 55000,
              'High Salary' : 0}
salary_55k
```

```python
# turn dictionary of scraped info into a dataframe for each filtered salary results. 
# then will need to concat all salary df's together and save as csv file
df_55 = pd.DataFrame(salary_55k)
df_55.head()
```
